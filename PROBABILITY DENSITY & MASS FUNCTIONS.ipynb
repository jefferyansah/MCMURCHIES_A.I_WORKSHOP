# DATA DISTRIBUTIONS 

(NOTE .PDF MEANS PROB DISTRIBUTION FUNCTION)  
(NOTE .PMF MEANS PROB MASS FUNCTION)  
  
THESE ARE ALL A WAY TO VISUALIZE AND MEASURE THE CHANCE OF A GIVEN DATA SET OCCURING

## UNIFORM DISTRIBUTION
FLAT CONSTANT PROBABILITY OF FINDING A VALUE IN GIVEN RANGE 

```python
import numpy as np
import matplotlib.pyplot as plt

values = np.random.uniform(-10.0, 10.0, 100000)
plt.hist(values, 50)
plt.show()
```

- WHAT WOULD THE PROBABILITY DISTRIBUTION FUNCTION LOOK LIKE? 
- IT WOULD BE ZERO UP TO -10 THEN IT WOULD BE A FLAT LINE THEN DROPPING TO ZERO AT 10 

# NORMAL / GAUSSIAN 
VISUALIZE PROBABILITY DENSITY FUNCTION  
Arrange from -3 to 3 with incriments of 0.001 for x values  
y value will be norm.pdf of corresponding x value 

```python
from scipy.stats import norm 
import matplotlib.pyplot as plt

x = np.arange(-3,3, 0.001)
plt.plot(x, norm.pdf(x))
```

## Random numbers with normal distribution

Lets generate some random numbers with a normal distribution 'mu' is the desired mean, sigma is the standard deviation. 


```python
import numpy as np
import matplotlib.pyplot as plt 

mu = 5.0 # CENTER
sigma = 2.0 # SPREAD

values = np.random.normal(mu, sigma, 10000)
plt.hist(values, 50)
plt.show()
```

# EXPONENTIAL PDF / "POWER LAW" 

```python
from scipy.stats import expon
import matplotlib.pyplot as plt

x=np.arange(0,10, 0.0001)
plt.plot(x, expon.pdf(x))
```

# BINOMIAL PROBABILITY MASS FUNCTION

DEALS WITH DISCRETE DATA

```python
from scipy.stats import binom
import matplotlib.pyplot as plt
n, p = 10, 0.5  # shape 
x = np.arange(0,10, 0.0001)
plt.plot(x, binom.pmf(x,n,p))
```

# POISSON PROBABILITY MASS FUNCTION
EXAMPLE MY SITE GETS 500 HITS PER DAY, WHATS THE PROBABILITY OF GETTING 550?   
SLIGHTLY DIFFERENT FROM NORMAL

```python
from scipy.stats import poisson 
import matplotlib.pyplot as plt

mu = 500 # mean
x = np.arange(400,600, 0.5)
plt.plot(x, poisson.pmf(x,mu))
```

SO PROBABILITY OF GETTING 550 IS ONLY 0.02%
